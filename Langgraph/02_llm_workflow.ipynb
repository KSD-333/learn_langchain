```python
from langchain_google_genai import GoogleGenerativeAI
from dotenv import load_dotenv
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
load_dotenv()  
```

```python
model = GoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.5)
```

```python
class LLMState(TypedDict):
    prompt: str
    response: str
```

```python
def llm(state: LLMState) -> LLMState:
    question = state['prompt']
    fined = f"Answer the following question concisely: {question}"
    mode_response = model.invoke(fined)
    state['response'] = mode_response
    return state
```

```python
graph = StateGraph(LLMState)

graph.add_node('llm',llm)

graph.add_edge(START,'llm')
graph.add_edge('llm',END)

work = graph.compile()
```

```python
result = work.invoke({'prompt':'What is LangGraph?'})
# Fixed: access the response from state, not .content
print(result['response'])
```
